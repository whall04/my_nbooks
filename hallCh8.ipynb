```python
import pandas as pd
import seaborn as sns
## matplotlib imported as plots were showing only object information but no visualizations. Using %matplotlib inline to force display. 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/Funkaphant/Downloads/avocado.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.query('year == 2015 and type == "conventional" and region == "Albany"')[['Date','Total Bags','Small Bags']].head()
```

```python
df_melted = pd.melt(df, id_vars='Date',
                      value_vars=['Total Bags','Small Bags'],
                      var_name='Bags', value_name='Count')
df_melted
```

```python
sns.relplot(data=df_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
df['Sales Volume'] = pd.qcut(df['Total Volume'], q=4, labels=('Poor', 'Modest', 'Good', 'Excellent'))
df.head()
```

```python
sns.countplot(x='year', hue='Sales Volume', data=df)
```

```python
df_melt2 = pd.melt(df, id_vars='year',
                      value_vars=['Small Bags','Large Bags','XLarge Bags'],
                      var_name='Bags', value_name='Count')
df_melt2
```

```python
sns.relplot(data=df_melt2, kind='line', x='year', y='Count', hue='Bags')
```
